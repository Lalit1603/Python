```python
import logging
logging.basicConfig(filename='even_parser.log',level=logging.DEBUG , format='%(asctime)s-%(message)s -%(levelname)s')
l=[4,5,6,7,8,9,0]
logging.info(l)
l1=[]
def ever_parser(l):
    try:
        
        logging.info("inside try block")
        for i in l:
            logging.info("insider for loop i =%d",i)
            if i%2==0:
                logging.info("checking the value of i")
                l1.append(i)
                logging.info("appending the value i = %d",i)
        return l1
    except Exception as e:
        logging.error(e)
        
```
